```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv("netflix.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data_2020 = data[data['release_year'] == 2020]


fig = px.bar(data_2020, x='type',
             title='shows Released in 2020',
             labels={'type': 'Type'}
          )
fig.show()
```

```python
import plotly.express as px

fig = px.pie(data, names='type',
             title='Distribution of Movie and TV Show Types',
             labels={'type': 'Type'})
fig.show()
```

```python
fig = px.scatter(data, x='release_year', y='show_id',
                 title='Scatter Plot of Release Year vs. Show ID',
                 labels={'release_year': 'Release Year', 'show_id': 'Show ID'})
fig.show()
```

```python
movies_data = data[data['type'] == 'Movie']


genre_counts = movies_data['listed_in'].str.split(', ').explode().value_counts()


top_5_genres = genre_counts.head(5)

fig = px.pie(top_5_genres, values=top_5_genres.values, names=top_5_genres.index,
             title='Top 5 Movie Genres', hole=0.4)
fig.show()
```

```python
data['before_after_2010'] = data['release_year'].apply(lambda year: 'Before 2010' if year < 2010 else 'After 2010')


grouped_data = data.groupby(['before_after_2010', 'type'])['show_id'].count().reset_index()

fig = go.Figure(data=[
    go.Bar(name='Movies', x=grouped_data[grouped_data['type'] == 'Movie']['before_after_2010'], y=grouped_data[grouped_data['type'] == 'Movie']['show_id']),
    go.Bar(name='TV Shows', x=grouped_data[grouped_data['type'] == 'TV Show']['before_after_2010'], y=grouped_data[grouped_data['type'] == 'TV Show']['show_id'])
])


fig.update_layout(barmode='group', title='Number of Shows Before and After 2010',
                  xaxis_title='Release Period', yaxis_title='Number of Shows')


fig.show()
```

```python
country_counts = data['country'].str.split(', ').explode().value_counts()


top_5_countries = country_counts.head(5)

fig = px.bar(top_5_countries,
             x=top_5_countries.index,
             y=top_5_countries.values,
             title='Top 5 Countries Producing Movies and TV Shows',
             labels={'x': 'Country', 'y': 'Number of Shows'})
fig.show()
```

```python
movies_data = data[data['type'] == 'Movie']

director_counts = movies_data.groupby('director')['show_id'].count().reset_index()


top_10_directors = director_counts.sort_values(by=['show_id'], ascending=False).head(10)


fig = px.bar(top_10_directors, x='director', y='show_id',
             title='Top 10 Directors by Number of Movies',
             labels={'director': 'Director', 'show_id': 'Number of Movies'})
fig.show()
```

```python
fig = px.scatter(data, y='director', x='show_id',
                 title='Scatter Plot of Director vs. Show ID',
                 labels={'release_year': 'Release Year', 'show_id': 'Show ID'})
fig.show()
```
